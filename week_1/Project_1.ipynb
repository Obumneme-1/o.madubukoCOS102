```python
# Class Project: Financial Calculations

def calculate_simple_interest():
    print("\n--- Simple Interest Calculation ---")
    P = float(input("Enter Principal (P): "))
    R = float(input("Enter Rate (R) in %: "))
    T = float(input("Enter Time (T) in years: "))
    
    # Formula: A = P * (1 + (R/100) * T)
    A = P * (1 + (R / 100) * T)
    print(f"The Total Amount (A) is: {A:.2f}")

def calculate_compound_interest():
    print("\n--- Compound Interest Calculation ---")
    P = float(input("Enter Principal (P): "))
    R = float(input("Enter Annual Rate (R) in decimal (e.g., 0.05 for 5%): "))
    n = int(input("Enter times interest compounded per year (n): "))
    t = float(input("Enter Time (t) in years: "))
    
    # Formula: A = P * (1 + R/n)**(n*t)
    A = P * (1 + R / n)**(n * t)
    print(f"The Total Amount (A) is: {A:.2f}")

def calculate_annuity_plan():
    print("\n--- Annuity Plan Calculation ---")
    PMT = float(input("Enter Periodic Payment (PMT): "))
    R = float(input("Enter Annual Rate (R) in decimal: "))
    n = int(input("Enter compounds per year (n): "))
    t = float(input("Enter Time (t) in years: "))
    
    # Formula: A = PMT * (((1 + R/n)**(n*t) - 1) / (R/n))
    numerator = (1 + R / n)**(n * t) - 1
    denominator = R / n
    A = PMT * (numerator / denominator)
    print(f"The Future Value of Annuity (A) is: {A:.2f}")

# Main execution loop
while True:
    print("\nSelect an option:")
    print("1. Simple Interest")
    print("2. Compound Interest")
    print("3. Annuity Plan")
    print("4. Exit")
    
    choice = input("Enter choice (1-4): ")
    
    if choice == '1':
        calculate_simple_interest()
    elif choice == '2':
        calculate_compound_interest()
    elif choice == '3':
        calculate_annuity_plan()
    elif choice == '4':
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```
