```python
# Project IV: Izifin Technology ATR Calculator

# 1. Take inputs from the user
try:
    experience = int(input("Enter years of experience: "))
    age = int(input("Enter age of the staff: "))

    # 2. Determine ATR based on conditions
    if experience > 25 and age >= 55:
        atr = 5600000
    elif experience > 20 and age >= 45:
        atr = 4480000
    elif experience > 10 and age >= 35:
        atr = 1500000
    else:
        # This covers staff with < 10 years experience and < 35 years of age
        atr = 550000
    
    # 3. Output the result
    print(f"\nThe Annual Tax Revenue (ATR) for the staff is: N{atr:,}")

except ValueError:
    print("Invalid input! Please enter numerical values for age and experience.")
```
